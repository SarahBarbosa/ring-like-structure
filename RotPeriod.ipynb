```python
import numpy as np
import scipy.signal as sps

def sigma_clip(x, nsigma=3):
    """
    Realiza a técnica de recorte sigma (sigma clipping) em dados unidimensionais.

    A técnica de recorte sigma é usada para remover valores discrepantes
    (outliers) em um conjunto de dados assumindo que os dados seguem uma
    distribuição Gaussiana (Normal). Essa função itera através dos dados
    e recorta valores que estão além de um determinado número de desvios padrão
    (sigma) a partir da mediana.

    Args:
        x (array): Array de dados a serem analisados. Supõe-se que os dados
                   sigam uma distribuição Gaussiana em 1D.
        nsigma (float): Número de desvios padrão para recortar. Padrão é 3.

    Returns:
        newx (array): Array de dados após o recorte dos valores discrepantes.
        mask (array): Máscara booleana indicando os valores que foram mantidos
                      após o recorte.
    """
    m = np.ones(len(x)) == 1
    newx = x * 1
    oldm = np.array([False])
    while sum(oldm) != sum(m):
        oldm = m * 1
        sigma = np.std(newx)
        m &= np.abs(np.median(newx) - x) / sigma < nsigma
        newx = x[m]
    return m

def filter_sigma_clip(x, y, nsigma=3, window_length=49, polyorder=3):
    """
    Realiza o recorte sigma em uma curva de luz utilizando um filtro de
    Savitzky-Golay.

    Esta função aplica um filtro de Savitzky-Golay para suavizar os dados de
    entrada e, em seguida, remove valores discrepantes utilizando a técnica de
    recorte sigma. O filtro de Savitzky-Golay é um método de suavização de dados
    que ajusta polinômios em janelas móveis dos dados.

    Args:
        x (array): Array de dados do eixo x.
        y (array): Array de dados do eixo y.
        nsigma (float, opcional): Número de desvios padrão para recortar.
                                  Padrão é 3.
        window_length (int, opcional): Comprimento da janela do filtro.
                                       Deve ser um número ímpar. Padrão é 49.
        polyorder (int, opcional): Ordem do polinômio usado no filtro de
                                   Savitzky-Golay. Padrão é 3.

    Returns:
        smooth (array): Array de dados suavizados.
        mask (array): Máscara booleana indicando os valores que
                      foram mantidos após o recorte.
    """
    smooth = sps.savgol_filter(y, window_length, polyorder)
    resids = y - smooth
    mask = sigma_clip(resids, nsigma=nsigma)
    return mask, smooth
```

# Plotting the Light Curve of a Rotating Star

```python
lc = lk.search_lightcurve("HD 101").download()
_, ax = plt.subplots()
lc.plot(ax = ax);
```

```python
# Converte arrays mascarados em arrays numpy regulares
flux = np.array(lc.flux.value)
flux_err = np.array(lc.flux_err.value)
time = np.array(lc.time.value)

# Calcula a mediana para normalização
med = np.median(flux)

# Realiza um recorte sigma inicial para remover grandes valores discrepantes
m = sigma_clip(flux / med - 1, nsigma=6)
x, y, yerr = time[m], flux[m] / med - 1, flux_err[m] / med

# Realiza um recorte sigma utilizando um filtro de Savitzky-Golay para suavização
mask, smooth = filter_sigma_clip(x, y, nsigma=3, window_length=199)
time, flux, flux_err = x[mask], y[mask], yerr[mask]
```

```python
_, ax = plt.subplots()
ax.scatter(lc.time.value, lc.flux.value/med-1, s=1, label="Outliers", c="slategrey")
ax.scatter(time, flux, s=1, label="Clipped", c="tab:blue")
ax.plot(x, smooth, label="Smoothed light curve", lw=1, c="darkblue")
ax.set(xlabel="Time [d]", ylabel="Normalized Flux", ylim=(-.005, .005))
plt.legend();
```

# Plotting the Periodogram of a Rotating Star

```python
pg = lc.to_periodogram(maximum_period=100)
period = pg.period_at_max_power.value

_, ax = plt.subplots()
pg.plot(view='period', ax = ax, lw=1)
ax.legend([f'Period at Max Power: {period:.2f} days']);
```

# Rotation Period for our F-stars

```python
url = "https://raw.githubusercontent.com/SarahBarbosa/ring-like-structure/main/data/Fstars_names.txt"
stars = pd.read_csv(url, header=None, names=["Name"])
stars["RotP"] = float("nan")

for i, star in enumerate(stars["Name"]):
    try:
        lc = lk.search_lightcurve(star).download()
        pg = lc.to_periodogram(maximum_period=100)
        period = pg.period_at_max_power.value
        stars.at[i, "RotP"] = period
    except Exception as e:
        print(f"An error occurred for star {star}: {e}")

print("Process completed.")
```

```python
stars.to_csv("Fstars_with_periods.csv", index=False)
```
